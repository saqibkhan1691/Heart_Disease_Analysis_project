```python
import pandas as pd
```

```python
df = pd.read_csv("Heart_new2.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check Null Values

df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
# Fix Yes/No Consistency

yes_no_cols = [
    "HeartDisease", "Smoking", "AlcoholDrinking",
    "Stroke", "DiffWalking", "PhysicalActivity", "Diabetic"
]

for col in yes_no_cols:
    df[col] = df[col].str.strip().str.capitalize()
```

```python
for col in yes_no_cols:
    print(col, df[col].unique())
```

```python
# Export the cleand dataset

df.to_csv("heart_data_clean.csv", index=False)
```
